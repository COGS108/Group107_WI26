## Abstract

Online multiplayer games rely on sustained player engagement in order to justify continued support, server upkeep, and ongoing updates. This study investigates whether an early decline in player activity can serve as a measurable predictor of long term player collapse within Steam multiplayer games. Using a dataset of over 612,000 monthly observations across different Steam games from July 2012 to September 2025, we analyzed historical trends in average and peak concurrent player counts. Our central independent variable was whether a game experienced a decline of at least 50% in average monthly players within 12 months of reaching its peak player count. The dependent variable was defined as a sustained collapse in player engagement, operationalized as falling to 5% or less of the game’s peak concurrent players within the following 24 months. 

Data analysis showed a substantial right-skew in player count distributions, prompting log transformations to normalize the data to offset the larger games. We then trained a K-Nearest Neighbors classifier using log-transformed engagement variables to evaluate whether early engagement declines were associated with eventual collapse. 

Model results achieved an overall accuracy of around 66%, with stronger performance finding games that remained active than those that collapsed. While the results suggest that early engagement declines contain some predictive information about long-term player outcomes, the relatively low precision and recall for collapsed games indicates that player decline alone is not sufficient to fully explain long-term engagement trajectories. This highlights the importance of understanding the complexity of what leads a game to sustained collapse, and to accurately predict their fate would require analyzing variables such as update frequency, genre, monetization marketing, and community interaction.

## Research Question

Does a decline of 50% or more in average monthly players within the first 12 months after a game’s peak significantly predict whether a Steam game will experience a sustained collapse in player engagement, defined as falling to 5% of its peak concurrent players, within the following 24 months?

## Background and Prior Work

Online multiplayer games depend on continued player engagements to justify the ongoing costs of server infrastructure, moderation, live updates and development support. These games require an active player base, and while engagement declines, companies must decide whether maintaining servers will be worth the commitment and resources used. There is yet to be a standardized, data driven benchmark that defines when a game has reached the point of sustained collapse. As a result, the threshold at which a game is considered no longer viable remains unclear. The lack of transparency creates uncertainty for players who are investing time and money without knowing how stable the game’s future truly is. 

Companies commonly consider several measurable factors when seeing whether to continue supporting certain titles. These can include concurrent player counts, monthly active users, retention rates, and the cost of maintaining development relative to revenue. Among these observable traits, player count is the most visible and consistently available metric, especially on platforms such as Steam. By tracking peaks and declines in concurrent players over time and often identify sharp falloffs following launch periods or major updates. In several live service cases, including MultiVersus, there is a noticeable pattern of an early peak followed by a rapid drop in players. While some decline after launch is expected, it is unclear at what point that decline signals long term collapse rather than normal market stabilization.

Existing research has approached this issue by modeling player retention and churn instead of defining a concrete collapse threshold. In Improved Retention Analysis in Freemium Role Playing Games by Jointly Modelling Players’ Motivation, Progression and Churn, the authors show that declining progression and reduced engagement are strong indicators that players are likely to stop playing altogether. Similarly, A Data Imputation Strategy to Enhance Online Game Churn Prediction, Considering Non Login Periods demonstrates that extended inactivity is a reliable predictor of permanent disengagement when properly modeled. These studies establish that measurable declines in engagement precede churn at the individual level. However, they do not define a population level benchmark that signals when a game as a whole is likely to experience sustained collapse. This gap motivates the present study, which tests whether a substantial early decline in average monthly players can serve as a measurable predictor of long term engagement deterioration across Steam multiplayer games.

^ Karmakar, Bikram, et al. “Improved retention analysis in freemium role-playing games by jointly modelling players’ motivation, progression and churn.” Journal of the Royal Statistical Society Series A: Statistics in Society, vol. 185, no. 1, 11 Aug. 2021, pp. 102–133, https://doi.org/10.1111/rssa.12730.

Each observation would represent a game-month, allowing us to track how player counts change over time and compare trajectories between games that remain stable and games that experience sustained collapse. Because the hypothesis focuses specifically on whether an early and substantial decline in player activity predicts long-term collapse, the most critical variables are those directly related to player counts and timing.

The dataset would include monthly average concurrent players, monthly peak concurrent players, each game’s historical peak player count, and an indicator for whether the game has shut down, along with the shutdown date if applicable. From these core variables, we would construct derived measures such as the percentage decline from the historical peak, whether a game experienced a decline of at least 50 percent within 12 months of reaching its peak, and whether the game later fell to 5 percent or less of its historical peak within 24 months. These constructed variables directly operationalize the independent variable of early substantial decline and the dependent variable of sustained collapse, making the dataset sufficient to test the central hypothesis.

While this limits our ability to directly evaluate financial viability or company decision making, it does not prevent us from testing whether early substantial declines in player activity are statistically associated with long-term engagement deterioration. ### KNN Classifier Model Analysis

In order to address our research question seeking to predict sustained collapse of player engagement, we used a KNN (K-Nearest Neighbors) classifier. This particular machine-learning algorithm is distance based so its quite sensative to the magnitude and scale of the input data. We employed standard scaling and applied it to our inputs (numerical) prior to fitting the model. This was done in order to make sure that the variables with greater absolute values didn't end up disproportionately skewing the distance calculations. For the analysis, we used a parameter of `n_neighbors = 10` for our KNN classifier. We used this particular value to make sure that our predictions would be based on a broader group of similar games. This would, therefore, reduce noise and smooth out potential for outliers that a smaller neighborhood might initialize. As we will see below, because this algorithm relies on the "majority vote" of 10 neighbors, the high number of games that are collapsing in the space often outnumbered the stable games in the neighborhoods. This lead to a low recall (0.27) for our model concerning the stable class. This isn't necessarily a case for worry because it makes sense that most games made would not be sustainable or "succeed" over the longterm by the industry's terms.

```python
#---------------Helper Functions---------------
#Function to identify if a game is dead/alive
def target_identifier(df_game):
    #-----Identify if the game is dead/alive-----
    peak_max = df_game['peak_players'].max() #Find the highest peak_players of each game
    curr = df_game['peak_players'].iloc[-1] #Find the most recent peak_players for the current game
    five_percent = peak_max * 0.05 #Find what 5% of the peak is
    status = 0 if ( curr  <= five_percent ) else 1 #Compare the most recent entry and five percent
                                                              #The game is considered Dead (0) if the current peak is <= 5 percent of the max peak
                                                              #The game is considered Alive (1) if the current peak is > 5 percent of the max peak
    df_game['Game_Status'] = status
    return df_game

#Function to identify if a game's avg player count decreased by half
def fifty_percent_decline(df_game):
    max_avg = df_game['avg_players'].max() #Find the highest avg_players of each game
    curr = df_game['avg_players'].iloc[-1] #Find the most recent avg_players for the current game
    status = True if ( curr <= max_avg/2 ) else False #Compare the most recent entry and 50% of max_avg, 
                                                      #True if the game avg deacreased by at least 50%. 
                                                      #False if the game avg deacreased by less than 50% 
    df_game['Declined_Fifty_Percent'] = status
    return df_game

#------------Load/Initialize Dataset-----------
df = pd.read_csv('data/02-processed/steamcharts.csv') #Load initial dataframe used in previous checkpoint
df['Declined_Fifty_Percent'] = True #Initialize a new categorical boolean column that identifies if the game decline by fifty percent. NOTE:
df['Game_Status'] = 'Dead' #Initialize a new categorical column that determines if the game is dead or alive. NOTE:
df = df.groupby('name').apply(target_identifier).reset_index(drop=True)
df = df.groupby('name').apply(fifty_percent_decline).reset_index(drop=True)

df_log_transform = df[['avg_players', 'peak_players', 'gain_percent', 'name', 'Game_Status', 'Declined_Fifty_Percent']] #Create a new dataframe that contains relevant data to transform

#----------------Transform Data----------------
#if (df['peak_players'] == 0).any(): print('There are 0s') #Checks for if there are 0s in peak_players
df_log_transform['peak_players'] = np.log(1 + df['peak_players']) #Log transform peak_players to normalize distribution, adding 1s since there were some entries with 0

#if (df['avg_players'] == 0).any(): print('There are 0s') #Checks for if there are 0s in avg_players
df_log_transform['avg_players'] = np.log(1 + df['avg_players']) #Log transform avg_players to normalize distribution, adding 1s since there were some entries with 0

#NOTE: We are not doing a log transformation on gain_percent since there are negative values and a log would not work on negative values
```

```python
#-----------------Correlations-----------------
corr = df_log_transform.drop(['name', 'Game_Status', 'Declined_Fifty_Percent'],axis=1).corr() #Observe the correlations between numeric data 

#----------Build a Pipeline and Model----------
#pipe = Pipeline([('scale', StandardScaler()), ('selector', 'passthrough'), ('svm', SVR())])
X = df_log_transform[['avg_players', 'peak_players', 'gain_percent']] #Numeric data to test on
y = df_log_transform['Game_Status'] #Target variable to test for

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42) #Split the data

pipe = Pipeline([('scale', StandardScaler()), ('knn', KNeighborsClassifier(n_neighbors=10))]) #Create a pipeline that will scale the data and use KNN
pipe.fit(X_train,y_train) #fit to the training data
results = pipe.predict(X_test) #predict the results based on the trained data


#OLS Model: NOTE: OLS is not the best model. The resulting R-squared was low, so the model did not fit the data the best
#avg_player_model = smf.ols(formula = 'Game_Status ~ avg_players + peak_players + gain_percent', data=df_log_transform) 
#results_avg_player = avg_player_model.fit()

```

#### Model Parameters

As noted prior, our KNN classifier model relies on the "majority vote" of 10 neighbors (`n_neighbors = 10`), the high number of games that are collapsing in the space often outnumbered the stable games in the neighborhoods. This lead to a low recall (0.27) for our model concerning the stable class which we will take a more in-depth look at in a moment.


#### Prediction Features
For our model, we used the core engagement metrics (`avg_players`, `peak_players`, and `gain_percent`) that were tracked in our data set to make a prediction of the games mortality. These predictions were split into class 1 representing the "dead/collapse" and the class 0 which represent the "Sstable" games. We used the engagement metrics mentioned prior but in the form of their log-transformed versions. We made sure to do this because it significantly helps in reducing the bias of hugely successful games, which would be outliers, by normalizing the extreme variance in the population. In prior analysis, we identified this to be a common characteristic of steam game populations. All in all,  class 1 & 0 seperations were determined based on our hypothesized early decline metric of 50% in the `avg_players` count from the given month where there was the highest max average (`max_avg`)(see research question). 

#### Accuracy
The overall accuracy of the model was 66% (0.66) on the test set of 306,133 different observations. This certainly indicates a greater predictive power than chance, however, we can get a better picture of the precision of the model by analyzing the recall, precision, and F1-scores. For the predictions of a games sustained collapse (class 1) the model seemed to preform quite well with a recall of 0.84. This means that it managed to catch 84% of the games that were actually "dead/collapsed" and had a precision score of 0.72. The resulting F1-score of 0.78 highlights that the model has strong reliability when it comes to predicting a given games collapse. For the games that were percieved as stable (class 0) the model seemed to struggle when it comes to identifying such. There was a precision of 0.43 and a percievably low recall of 0.27 which resulted in a corresponding F1-score of 0.33. Thus, the model seemed to frequently find itself misclassifying stable games as collapsing ones. This leads us into the model's drawbacks.

#### Limitations

For our model, the main drawback is that it faces heavy imbalances in class. The test data set we used contains about 211,189 collapsing instances while there are 94,844 stable instances of games. This can lead to inaccuracy in prediction because the KNN model classifies based on the "majority vote" of the nearest neighbors (the neighborhood). Because there is such a high density of games that considered "dead/collapsed" in the feature space it leads to the model aggressively over-predicting class 1 (collapsed) games.

In addition to this, we faced a notable limitation in our feature set. The model lacks the context beind the numbers of why these percieved sustained collapses occur. This is because we used `avg_players`, `peak_players`, and `gain_percent` as our primary metrics of prediction which do not offer such insight. Variables such as the means of monetization or lacktherof, genre, or game update frequency could be missing factors that could help in distinguishing whether the stabilization of players after a post-launch is normal rather than the death of game indicated by it's decline. Having such information would help in making more accurate predictions, and would be of more use to developers and creators who are looking for factors that influence a games sustained "success" or "collapse."

Because the player counts across Steam games were extremely right-skewed, due to only a small number of games reaching greater player populations compared to smaller communities, we adapted to use percentage-based thresholds rather than raw player counts to define collapse. These, along with log transformations were applied to player count variables to reduce the influence of larger games and make it accessible for the model. 

However, there were several limitations to our study that could explain why the predictive results were only moderate. We decided to only analyze player count, so possible confounding variables for contextualizing the peaks/declines could heavily impact the analysis of our data. Factors such as player retention, genre, update schedule, monetization, or community feedback were not included which all are important to analyzing a game's possible collapse. Not only this, but the dataset we accessed could’ve been affected over time by different Steam updates on data collectionDespite these limitations, our findings still provide useful insights to the lifecycle of online multiplayer games overall. Early declines in player activity is one indicator that a game may struggle to maintain long-term engagement, but they are only one part of a much larger system. 

Understanding the full picture of predicting a games collapse would require combining our findings, with additional engagement metrics like playtime or community activity as well. By expanding the dataset through incorporating differing variables, future work could examine how update frequency, monetization marketing, or review sentiment could save a game from said collapse. Overall, this project demonstrates that player population data can provide meaningful insights into a game's health, but predicting the health long term requires a more complex analysis. Early engagement decline may serve as a warning sign, but it does not fully determine a game’s future trajectory.
